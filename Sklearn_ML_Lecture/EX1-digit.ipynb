# Digit

# Load dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits
```

```python
dataobj = load_digits()
```

```python
X = pd.DataFrame(dataobj.data, columns=dataobj.feature_names)
X
```

```python
y = dataobj.target
y
```

```python
df = X
df['target'] = y
df
```

# ploting picture

```python
number=8
n_pic=5

fig, ax = plt.subplots(1, n_pic, figsize=(20, 10))
ax = ax.ravel()
for e in range(n_pic):
    ax[e].matshow(df[df.target == number].drop('target', axis=1).to_numpy()[e].reshape(-1, 8))
```

# Training model using pipline
- Spliting data
- Standardized data
- PCA with 90% cumulative explained variance
- Train model wih SVM

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC

from sklearn import metrics
```

```python
## Spliting dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.7, random_state=1, stratify=y)
```

```python
## Making pipline
scl = StandardScaler()
pca = PCA(n_components=.9)
clf = SVC(kernel='rbf', C=100, gamma=.001, random_state=10)

pipe = Pipeline([
    ('scale', scl),
    ('pca', pca),
    ('clf', clf)
])

pipe
```

```python
## fitting model
pipe.fit(X_train, y_train)
```

```python
print(f"Train accuracy = {pipe.score(X_train, y_train)}")
```

# Cross validation

```python
from sklearn.model_selection import cross_validate
score = cross_validate(estimator=pipe, X=X_train, y=y_train, cv=10, scoring=['accuracy'], n_jobs=-1)
score
```

```python
print(f'Validation Accuracy = {np.mean(score["test_accuracy"])}')
```

# Learning Curve

```python
from sklearn.model_selection import learning_curve
```

```python
train_size = np.linspace(.1, 1, 10)
train_size
```

```python
train_size, train_score, val_score = learning_curve(
    estimator=pipe,
    X=X_train,
    y=y_train,
    train_sizes=train_size,
    cv=10,
    n_jobs=-1
)
```

```python
df_train_score = pd.DataFrame(train_score, index=train_size)
df_train_score
```

```python
sum_train = df_train_score.aggregate(['mean', 'std'], axis =1)
sum_train
```

```python
df_val_score = pd.DataFrame(val_score, index=train_size)
df_val_score
```

```python
sum_val = df_val_score.aggregate(['mean', 'std'], axis =1)
sum_val
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(sum_train['mean'], marker='o')
ax.plot(sum_val['mean'], marker='o')
ax.set_xlabel('Number of Samples')
ax.set_ylabel('Accuracy score')
ax.set_title('Learing Curve of model')
```

# Validation curve

```python
from sklearn.model_selection import validation_curve
```

```python
pipe.get_params()
```

```python
param_range = [.0001, .001, .01, .1, 1, 10, 100, 1000]

train_score, val_score = validation_curve(
    estimator=pipe,
    X=X_train,
    y=y_train,
    cv=10,
    n_jobs=-1,
    param_name='clf__C',
    param_range=param_range
)
```

```python
mean_train_score = np.mean(train_score, axis=1)
mean_val_score = np.mean(val_score, axis=1)
```

```python
## plotting
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(param_range, mean_train_score, marker='o', label='Train Dataset')
ax.plot(param_range, mean_val_score, marker='o', label='Validation Dataset')
ax.set_xscale('log')
ax.set_xlabel('C')
ax.set_ylabel('Accuracy Score')
ax.set_title('Validation Curve')
ax.legend();
```

# Grid Search

```python
pipe.get_params()
```

```python
## creating set of hyper parameter
param_range = [.0001, .001, .01, .1, 1, 10, 100, 1000]
set1 = {
    'clf__C':param_range,
    'clf__gamma':param_range,
    'clf__kernel':['linear']
}

set2 = {
    'clf__C':param_range,
    'clf__gamma':param_range,
    'clf__kernel':['rbf']
}

set3 = {
    'clf__C':param_range,
    'clf__gamma':param_range,
    'clf__kernel':['poly']
}
param_grid = [set1, set2, set3]
```

```python
## grid scearch cross validation
from sklearn.model_selection import GridSearchCV

gs = GridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    scoring='accuracy'

)
gs
```

```python
gs.fit(X_train, y_train)
```

```python
re = pd.DataFrame(gs.cv_results_)
re.sort_values('rank_test_score')
```

```python
## best score and hyper parameter
print(f'Best accuracy : {gs.best_score_}')
print(f'Best Hyperprameter : {gs.best_params_}')
```

```python
## refitting with best hyper parameter
print(f'Refitting model : {gs.refit}')
```

```python
## evaluate model with testing data
y_pred = gs.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
## confusion metrix
cm = metrics.confusion_matrix(y_test, y_pred)
fig = metrics.ConfusionMatrixDisplay(cm)
fig.plot()
```

# Randomied Search

```python
## creating set of hyper parameter
param_range = [.0001, .001, .01, .1, 1, 10, 100, 1000]
set1 = {
    'clf__C':param_range,
    'clf__gamma':param_range,
    'clf__kernel':['linear']
}

set2 = {
    'clf__C':param_range,
    'clf__gamma':param_range,
    'clf__kernel':['rbf']
}

set3 = {
    'clf__C':param_range,
    'clf__gamma':param_range,
    'clf__kernel':['poly']
}
param_grid = [set1, set2, set3]
```

```python
## grid scearch cross validation
from sklearn.model_selection import RandomizedSearchCV

gs = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=param_grid,
    cv=5,
    n_jobs=-1,
    scoring='accuracy',
    n_iter=50

)
gs
```

```python
gs.fit(X_train, y_train)
```

```python
re = pd.DataFrame(gs.cv_results_)
re.sort_values('rank_test_score').head(5)
```

```python
## best score and hyper parameter
print(f'Best accuracy : {gs.best_score_}')
print(f'Best Hyperprameter : {gs.best_params_}')
```

```python
## refitting with best hyper parameter
print(f'Refitting model : {gs.refit}')
```

```python
## evaluate model with testing data
y_pred = gs.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
## confusion metrix
cm = metrics.confusion_matrix(y_test, y_pred)
fig = metrics.ConfusionMatrixDisplay(cm)
fig.plot()
```
