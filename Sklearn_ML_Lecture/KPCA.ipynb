# Kernel Principle Component Analysis

# Load library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import  datasets

%matplotlib inline
%config InlineBackend.figureformat = 'retina'
```

```python
# Plot function
def plot(x, y, type_d='normal'):
    fig, ax = plt.subplots(figsize = (20, 10))
    ax.scatter(x[y == 0, 0], x[y == 0, 1], color='red', marker='o', alpha=.5)
    ax.scatter(x[y == 1, 0], x[y == 1, 1], color='blue', marker='d', alpha=.5)
    
    if type_d == 'normal':
        ax.set_title('Original Data')
        ax.set_xlabel('Axis 1')
        ax.set_ylabel('Axis 2')
        
    elif type_d == 'pca':
        ax.set_title('PCA Data')
        ax.set_xlabel('Axis 1')
        ax.set_ylabel('Axis 2')
        
    else:
        ax.set_title('KPCA Data')
        ax.set_xlabel('Axis 1')
        ax.set_ylabel('Axis 2')
    
```

# Load data

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, random_state=123)
plot(X, y)
```

# PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scl = StandardScaler()
X_s = scl.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_s)
plot(X_pca, y, type_d='pca')
```

# KPCA

```python
from sklearn.decomposition import KernelPCA
```

```python
kpca = KernelPCA()
kpca.get_params()
```

```python
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=3)
X_kpca = kpca.fit_transform(X_s)
plot(X_kpca, y, type_d='kpca')
```

```python
gamma = [.001, .01, .1, 1, 10, 100, 1000]
for g in gamma:
    kpca = KernelPCA(n_components=2, kernel='rbf', gamma=g)
    X_kpca = kpca.fit_transform(X_s)
    plot(X_kpca, y, type_d='kpca')
    print(f'gamma = {g}')
```

# Concentric circle

```python
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=1000, random_state=123, noise=.1, factor=.2)
plot(X, y)
```

```python
scl = StandardScaler()
X_s = scl.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_s)
plot(X_pca, y, type_d='pca')
```

```python
gamma = [.001, .01, .1, 1, 10, 15, 20]
for g in gamma:
    kpca = KernelPCA(n_components=2, kernel='rbf', gamma=g)
    X_kpca = kpca.fit_transform(X_s)
    plot(X_kpca, y, type_d='kpca')
    print(f'gamma = {g}')
```
