# TSNE

# Load library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import  datasets

%matplotlib inline
%config InlineBackend.figureformat = 'retina'
```

```python
# Plot function
def plot(x, y, type_d='normal'):
    fig, ax = plt.subplots(figsize = (20, 10))
    ax.scatter(x[y == 0, 0], x[y == 0, 1], color='red', marker='o', alpha=.5)
    ax.scatter(x[y == 1, 0], x[y == 1, 1], color='blue', marker='d', alpha=.5)
    
    if type_d == 'normal':
        ax.set_title('Original Data')
        ax.set_xlabel('Axis 1')
        ax.set_ylabel('Axis 2')
        
    elif type_d == 'pca':
        ax.set_title('PCA Data')
        ax.set_xlabel('Axis 1')
        ax.set_ylabel('Axis 2')
        
    elif type_d == 'kpca':
        ax.set_title('KPCA Data')
        ax.set_xlabel('Axis 1')
        ax.set_ylabel('Axis 2')
        
    elif type_d == 'tsne':
        ax.set_title('TSNE Data')
        ax.set_xlabel('Axis 1')
        ax.set_ylabel('Axis 2')  
```

# Load Dataset

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=500, random_state=123)
plot(X, y)
```

# TSNE

```python
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
```

```python
scl = StandardScaler()
X_s = scl.fit_transform(X)
```

```python
tsne = TSNE(n_components=2, perplexity=5, learning_rate=10)
X_tsne = tsne.fit_transform(X_s)
plot(X_tsne, y, type_d='tsne')
```
